```python
fig = sns.countplot(data=df, x='Gender')
for bars in fig.containers:
    fig.bar_label(bars)
```

```python
SalesData_Gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

figure = sns.barplot(x='Gender', y='Amount', data=SalesData_Gender)

for bars in figure.containers:
    figure.bar_label(bars)
```

```python
fig = sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in fig.containers:
    fig.bar_label(bars)
```

```python
SaleData_Agegroup = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

figure = sns.barplot(data=SaleData_Agegroup, x='Age Group', y='Amount')
```

```python
df.columns
```

```python
NumberOfOrdersByStates = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(ascending=False, by='Orders')
plt.figure(figsize=(25, 6))
fx = sns.barplot(data=NumberOfOrdersByStates, x='State', y='Orders')
```

```python
AmounSpentByStates = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(25,6))

fy = sns.barplot(data=AmounSpentByStates, x='State', y='Amount')
```

```python
pk = sns.countplot(data=df, x='Marital_Status')

#plt.figure(figsize=(10,3))

for bars in pk.containers:
    pk.bar_label(bars)
```

```python
SalesData_MS = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sp = sns.barplot(data=SalesData_MS, x='Marital_Status', y='Amount', hue='Gender')
```

```python
plt.figure(figsize=(30,6))
gd = sns.countplot(data=df, x='Occupation')

for bars in gd.containers:
    gd.bar_label(bars)
```

```python
SalesgenbyOccupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
plt.figure(figsize=(30,6))
ds = sns.barplot(data=SalesgenbyOccupation, x='Occupation', y='Amount')
```
